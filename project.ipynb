```python
c1_vecc = np.array(df_class1['0'])
c2_vecc = np.array(df_class1['1'])
```

```python
c1_vecc
```

```python
def min_dis(a,b,r):
    return np.sum(np.abs(a-b)**r)**(1/r)

r_values = np.arange(1,11)
distances = [min_dis(c1_vecc,c2_vecc,r) for r in r_values]

plt.figure(figsize=(8,8))
plt.plot(r_values,distances,marker='o')
plt.xlabel('r')
plt.ylabel('Min distance')
plt.title('Min distance between 2 feature vector')
plt.xticks(r_values)
plt.grid()
plt.show()
```
