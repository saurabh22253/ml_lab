```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
# Load the dataset
df1 = pd.read_csv('English_Abstractive.csv')

# Split the dataset into two classes based on 'Judgement Status'
df_class1 = df1[df1['Judgement Status'] == 0]
df_class2 = df1[df1['Judgement Status'] == 1]

# Calculate the mean vector for Class 2 (excluding the last column)
mean_class2 = np.mean(df_class2.iloc[:, :-1])
print(mean_class2)

# Calculate the mean vector for Class 1 (excluding the last column)
mean_class1 = np.mean(df_class1.iloc[:, :-1])
print(mean_class1)
```

```python
# Create a list to store mean values for each feature in Class 1
l1 = [] 
for i in range(len(df_class1.columns) - 1):
    mean_value = np.mean(df_class1.iloc[:, i])  # Calculate mean for each feature in Class 1
    l1.append(mean_value) 
mean_vecc1 = np.array(l1)  # Convert the list to a numpy array (mean vector for Class 1)
print(mean_vecc1)
```

```python
# Create a list to store mean values for each feature in Class 2
l2 = [] 
for i in range(len(df_class2.columns) - 1):
    mean_value = np.mean(df_class2.iloc[:, i])  # Calculate mean for each feature in Class 2
    l2.append(mean_value) 
mean_vecc2 = np.array(l2)  # Convert the list to a numpy array (mean vector for Class 2)
print(mean_vecc2)


# Calculate the Euclidean distance between the mean vectors of Class 1 and Class 2
distance_between_means = np.linalg.norm(mean_vecc1 - mean_vecc2)
print(distance_between_means)
```

```python
# Create a list to store standard deviation values for each feature in Class 1
l3 = [] 
for i in range(len(df_class1.columns) - 1):
    std_value = np.std(df_class1.iloc[:, i])  # Calculate standard deviation for each feature in Class 1
    l3.append(std_value) 
std_vecc1 = np.array(l3)  # Convert the list to a numpy array (standard deviation vector for Class 1)
std_vecc1
```

```python
# Create a list to store standard deviation values for each feature in Class 2
l4 = [] 
for i in range(len(df_class2.columns) - 1):
    std_value = np.std(df_class2.iloc[:, i])  # Calculate standard deviation for each feature in Class 2
    l4.append(std_value) 
std_vecc2 = np.array(l4)  # Convert the list to a numpy array (standard deviation vector for Class 2)
std_vecc2
```

```python
import matplotlib.pyplot as plt
```

```python
# Choose a feature (column) for histogram analysis
c1 = df_class1['0']

# Define the number of bins for the histogram
bins = 10

# Calculate the histogram data
hist, bin_edges = np.histogram(c1, bins=bins)

# Calculate mean and variance for the selected feature
mean_c1 = np.mean(c1)
var_c1 = np.var(c1)

# Plot the histogram for the selected feature
plt.figure(figsize=(8,8))
plt.hist(c1, bins=bins)
plt.xlabel('Feature 0 Value')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
# Select two feature vectors from Class 1 for Minkowski distance calculation
c1_vecc = np.array(df_class1['0'])
c2_vecc = np.array(df_class1['1'])

# Function to calculate Minkowski distance
def min_dis(a, b, r):
    return np.sum(np.abs(a - b) ** r) ** (1 / r)

# Calculate Minkowski distances for r values from 1 to 10
r_values = np.arange(1, 11)
distances = [min_dis(c1_vecc, c2_vecc, r) for r in r_values]

# Plot Minkowski distances as a function of r
plt.figure(figsize=(8,8))
plt.plot(r_values, distances, marker='o')
plt.xlabel('r')
plt.ylabel('Minkowski Distance')
plt.title('Minkowski Distance between 2 Feature Vectors')
plt.xticks(r_values)
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Separating features and labels
x = df1.iloc[:, :-1].values  # Feature vectors
y = df1['Judgement Status'].values  # Labels

# Splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initializing the kNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Training the classifier
knn.fit(x_train, y_train)
```

```python
# Testing the classifier
accuracy = knn.score(x_test, y_test)
print(f"Accuracy of kNN (k=3): {accuracy * 100:.2f}%")
```

```python
# Predicting the class labels for the test set
y_pred = knn.predict(x_test)

# Displaying the first 10 predictions alongside actual labels
for i in range(10):
    print(f"Predicted: {y_pred[i]}, Actual: {y_test[i]}")
```

```python
accuracies = []
k_values = range(1, 12)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    accuracies.append(knn.score(x_test, y_test))

# Plotting the accuracy
plt.figure(figsize=(8, 8))
plt.plot(k_values, accuracies, marker='o')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy of kNN Classifier with varying k')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")

# Precision, Recall, and F1-Score for multiclass
# 'macro' average calculates metrics for each label, and then takes the unweighted mean
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```
